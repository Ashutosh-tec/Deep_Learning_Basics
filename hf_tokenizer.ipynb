```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.pre_tokenizers import Whitespace
from tokenizers import normalizers
from tokenizers.normalizers import Lowercase
from tokenizers.trainers import BpeTrainer
```

```python
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
tokenizer.pre_tokenizer = Whitespace()

normalizer = normalizers.Sequence([Lowercase()])
tokenizer.normalizer = normalizer

trainer = BpeTrainer(special_tokens=["[GO]", "[UNK]", "[PAD]", "[EOS]"], vocab_size=5000, continuing_subword_prefix="##")
tokenizer.train_from_iterator(ds['text'], trainer=trainer)
```

```python
def custom_tokenizer(dataset, vocab_size):

  # Select every 7th sample from the dataset
  # Merge the 'train' split into a list of texts, then filter every 7th text
  texts = [dataset['train'][i]['text'] for i in range(0, len(dataset['train']), 7)]
  print(f"Number of selected samples: {len(texts)}")  # Should result in approximately 10 million samples
  # Define the BPE tokenizer with required configuration
  tokenizer = Tokenizer(BPE())

  # Configure normalizer and pre-tokenizer
  tokenizer.normalizer = normalizers.Sequence([Lowercase()])
  tokenizer.pre_tokenizer = Whitespace()

  # Special tokens
  special_tokens = ["[GO]", "[UNK]", "[PAD]", "[EOS]"]

  # Train the tokenizer on the selected texts with a vocabulary size of 5000
  trainer = BpeTrainer(vocab_size=vocab_size, special_tokens=special_tokens, continuing_subword_prefix="##")

  # Train the tokenizer on the selected dataset
  tokenizer.train_from_iterator(texts, trainer=trainer)
  return tokenizer




custom_tok = custom_tokenizer(ds, 32000)
```

We have custom_tok, hopper, gpt_tok and bert_tok

```python
texts = imdb['train']['text'] + imdb['test']['text']  # Combine train and test splits

# Function to count tokens for a dataset using a given tokenizer
def count_tokens_with_tokenizer(tokenizer, texts, is_hf_tokenizer=False):
    total_tokens = 0
    for text in texts:
        if is_hf_tokenizer:
            # Use HuggingFace tokenizers
            encoded = tokenizer.encode(text)
            total_tokens += len(encoded)
        else:
            # Use custom tokenizers (e.g., Tokenizer class)
            encoded = tokenizer.encode(text)
            total_tokens += len(encoded.tokens)
    return total_tokens

# Step 3: Count tokens for each tokenizer
custom_token_count = count_tokens_with_tokenizer(custom_tok, texts)
bert_token_count = count_tokens_with_tokenizer(bert_tok, texts, is_hf_tokenizer=True)
gpt2_token_count = count_tokens_with_tokenizer(gpt_tok, texts, is_hf_tokenizer=True)
hopper_token_count = count_tokens_with_tokenizer(hopper, texts)
```

```python
token_arr = [(custom_token_count, "Custom"), (bert_token_count, "Bert"), (gpt2_token_count, "GPT2"), (hopper_token_count, "Hopper")]
sorted(token_arr, key=lambda x: x[0])
```
